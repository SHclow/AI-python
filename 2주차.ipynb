```python
# 1. 변수 선언: 계산할 숫자를 변수에 담습니다.
number1 = 10
number2 = 3

# 2. 연산자 사용: 사칙연산 결과를 새로운 변수에 저장합니다.
result_add = number1 + number2      # 덧셈
result_sub = number1 - number2      # 뺄셈
result_mul = number1 * number2      # 곱셈
result_div = number1 / number2      # 나눗셈 (소수점 포함)
result_floor = number1 // number2   # 몫 (정수 나눗셈)
result_mod = number1 % number2      # 나머지

# 3. 출력문: 저장된 변수의 값을 화면에 보여줍니다.
print("첫 번째 숫자:", number1)
print("두 번째 숫자:", number2)
print("-" * 20)  # 구분선 (문자열 곱하기 연산자 활용)
print("더하기 (+):", result_add)
print("빼기 (-):", result_sub)
print("곱하기 (*):", result_mul)
print("나누기 (/):", result_div)
print("몫 (//):", result_floor)
print("나머지 (%):", result_mod)
```

```python
# 1. 입력 및 변수 선언: input()으로 받고 int()로 숫자로 바꿉니다.
# "문자열"을 입력받아 바로 "정수"로 변환하여 변수에 저장합니다.
number1 = int(input("첫 번째 숫자를 입력하세요: "))
number2 = int(input("두 번째 숫자를 입력하세요: "))

# 2. 연산자 사용: 사칙연산 결과를 새로운 변수에 저장합니다.
result_add = number1 + number2      # 덧셈
result_sub = number1 - number2      # 뺄셈
result_mul = number1 * number2      # 곱셈
result_div = number1 / number2      # 나눗셈 (소수점 포함)
result_floor = number1 // number2   # 몫 (정수 나눗셈)
result_mod = number1 % number2      # 나머지

# 3. 출력문: 저장된 변수의 값을 화면에 보여줍니다.
print("-" * 20)  # 입력과 결과를 구분하기 위한 선
print("입력한 숫자:", number1, ",", number2)
print("-" * 20)
print("더하기 (+):", result_add)
print("빼기 (-):", result_sub)
print("곱하기 (*):", result_mul)
print("나누기 (/):", result_div)
print("몫 (//):", result_floor)
print("나머지 (%):", result_mod)
```
